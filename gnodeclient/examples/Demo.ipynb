## Start the session

Import the gnodeclient package and start the session:

```python
import neo
from gnodeclient import session, Model

g = session.create(location="http://beta.g-node.org", username="demo", password="pass")
```

## Upload sample data

Read some sample data array (e.g. LFP trace or membrane potential):

```python
import numpy as np
import quantities as pq
import matplotlib.pyplot as plt

# read sample data (LFP) from file
with open('sample.dat', 'r') as f:
    line = f.readlines()[0]
    
line = line.split(" ")
for i in range(line.count('')):
    line.remove('')
line = [float(el) for el in line]

# plot the signal 
sampling_rate = 500.0 * pq.Hz
start = 0.0 * pq.ms
duration = start + len(line)/sampling_rate

ticks = np.arange(start, start + duration, 0.0 * pq.ms + 1/sampling_rate)
data = pq.Quantity(line, units="mV")

plt.plot(ticks, data)
```

Build a sample dataset (block) that binds previously read data array to a certain time segment:

```python
import uuid

unique_name = uuid.uuid1().hex[:10]
print "The unique sample name is %s" % unique_name

block = neo.Block(name="Sample block with %s" % unique_name)

segment = neo.Segment(name="Sample segment with %s" % unique_name)
segment.block = block
block.segments.append(segment)

signal_params = {
    'name': "Sample signal with %s" % unique_name,
    't_start': start,
    'sampling_rate': sampling_rate,
    'signal': data
}
signal = neo.AnalogSignal(**signal_params)
signal.segment = segment
segment.analogsignals.append(signal)
```

Save dataset on the server:

```python
block = g.set_all(block)
```

## Fetch saved data

Fetch this block later from the server:

```python
blocks = g.select(Model.BLOCK, {"name__icontains": unique_name})
print "%d blocks fetched" % len(blocks)

blocks[0].location == block.location
```

Plot fetched signal from the dataset:

```python
#signal = g.select(Model.ANALOGSIGNAL, {"name__icontains": unique_name})[0]
signal = blocks[0].segments[0].analogsignals[0]

ticks = np.arange(signal.t_start, signal.t_stop, 0.0 * pq.ms + 1/signal.sampling_rate)

# units are stored with the signal object
plt.xlabel("time [%s]" % signal.times.units.dimensionality.string)
plt.ylabel("voltage [%s]" % signal.units.dimensionality.string)

plt.plot(ticks, signal)
```

## Open data for a collaborator:

```python
user_account = "/api/v1/user/user/ed/"

g.permissions(block, [{"user": user_account, "access_level": 1}]) # read-only
```

Access shared data with another account:

```python
from datetime import datetime

e = session.create(location="http://beta.g-node.org", username="ed", password="pass")

query = {
    'date_created__month': datetime.today().month,
    'date_created__day': datetime.today().day,
    'max_results': 5
}
signal = e.select(Model.ANALOGSIGNAL, query)[0]

ticks = np.arange(signal.t_start, signal.t_stop, 0.0 * pq.ms + 1/signal.sampling_rate)

# units are stored with the signal object
plt.xlabel("time [%s]" % signal.times.units.dimensionality.string)
plt.ylabel("voltage [%s]" % signal.units.dimensionality.string)

plt.plot(ticks, signal)
```
